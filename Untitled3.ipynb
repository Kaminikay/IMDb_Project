**Subtask 3.6: Find the popular Genres **

1. Extract the first two genres from the genre column and store them in the new column genre_1 and genre_2 . Some of the movies contains only genre_1 so in that case put the value of genre_1 in the genre_2 column
2. Group the dataframe using genre_1 as the primary column and genre_2 as the secondary column
3. Find the 5 most popular combo of genres by finding the mean of the gross and store them in a new dataframe pop_genre.

```python
temp_genre=movies.genres.str.split('|',expand=True).iloc[:,0:2]
temp_genre.columns=['genre_1','genre_2']
temp_genre.genre_2.fillna(temp_genre.genre_1,inplace=True)
temp_genre
```

```python
movies=pd.concat([movies,temp_genre],axis=1)
movies
```

```python
# Group the dataframe by genre_1 and genre_2 as per gross_mean
movies.groupby(['genre_1','genre_2']).gross.mean().sort_values(ascending=False).head(5)
```
